```python
def primes(n):
    sieve = [True] * (n+1)
    for i in range(2, n//2 + 1):
        if not sieve[i]:
            continue
        for k in range(i * 2, n+1, i):
            sieve[k] = False
    return [i for i in range(2, n+1) if sieve[i]]
```

```python
primes(1000000)
```

```python
def primes(n):
    sieve = [True] * (n+1)
    for i in range(2, round(sqrt(n)) + 1): # changed n//2  to round(sqrt(n))
        if not sieve[i]:
            continue
        for k in range(i * i, n+1, i): # changed i*2 to i*i
            sieve[k] = False
    return [i for i in range(2, n+1) if sieve[i]]
```

```python
primes(100_000_000)
```
